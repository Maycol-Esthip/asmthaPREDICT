```python
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score, roc_curve, confusion_matrix
import matplotlib.pyplot as plt

# Balanceo de datos
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_trainset, y_trainset)

# Modelo mejorado
model = RandomForestClassifier(max_depth=15, max_leaf_nodes=20, n_estimators=200, random_state=42)
model.fit(X_train_balanced, y_train_balanced)

# Predicciones
y_pred = model.predict(X_trainset)

# Evaluación del modelo
print(classification_report(y_trainset, y_pred))

# AUC-ROC
y_probas = model.predict_proba(X_trainset)[:, 1]
auc_score = roc_auc_score(y_trainset, y_probas)
print(f'AUC-ROC: {auc_score}')

# Matriz de confusión
conf_matrix = confusion_matrix(y_trainset, y_pred)
print(conf_matrix)

# Curva ROC
fpr, tpr, _ = roc_curve(y_trainset, y_probas)
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (AUC = %0.2f)' % auc_score)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
# Importar las métricas necesarias para la Curva ROC y el cálculo del AUC
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import roc_auc_score
from sklearn.metrics import RocCurveDisplay
from sklearn.metrics import precision_recall_curve
from itertools import cycle
import matplotlib.pyplot as plt

# Obtener las probabilidades de la clase positiva del conjunto de entrenamiento
probs = model.predict_proba(X_trainset)
probs = probs[:, 1]  # Selecciona las probabilidades para la clase 1 (positiva)

# Crear y graficar la curva ROC usando las probabilidades predichas
RocCurveDisplay.from_predictions(
    y_trainset,
    probs,
    name="ROC Curve",
    color="darkorange",
)

# Graficar la línea de referencia para el nivel de azar (AUC=0.5)
plt.plot([0, 1], [0, 1], "k--", label="Chance level (AUC = 0.5)")

# Configuración del gráfico
plt.axis("square")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.rcParams["figure.figsize"] = (4, 4)  # Tamaño de la gráfica 4x4
plt.title("RandomForest ROC Curve (Train)")
plt.legend()
plt.show()
```
